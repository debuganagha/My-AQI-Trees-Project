The Air Quality Index (AQI) ranges in OpenWeather are typically classified into six categories, each indicating a different level of health concern:

    Good (0-50): Air quality is considered satisfactory, and air pollution poses little or no risk.
    Moderate (51-100): Air quality is acceptable; however, there may be a moderate health concern for people who are unusually sensitive to air pollution.
    Unhealthy for Sensitive Groups (101-150): Members of sensitive groups (such as children, the elderly, and people with respiratory conditions) may experience health effects. The general public is less likely to be affected.
    Unhealthy (151-200): Everyone may begin to experience health effects; members of sensitive groups may experience more serious health effects.
    Very Unhealthy (201-300): Health alert: everyone may experience more serious health effects.
    Hazardous (301-500): Health warning of emergency conditions. The entire population is likely to be affected.

```python
import matplotlib.pyplot as plt

# Calculate the frequency of each AQI value
aqi_counts = merged_data['AQI'].value_counts().sort_index()

# Plot the AQI frequency as a bar graph
plt.figure(figsize=(8, 6))
plt.bar(aqi_counts.index, aqi_counts.values, color='skyblue', edgecolor='black')
plt.title("Frequency of AQI Values", fontsize=14)
plt.xlabel("AQI Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(aqi_counts.index)  # Display AQI values on the x-axis
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting a histogram for CO values with a KDE line
plt.figure(figsize=(6, 6))
sns.histplot(merged_data['CO'], bins=30, color='lightgreen', edgecolor='black', kde=True)
plt.title("Distribution of CO Levels", fontsize=14)
plt.xlabel("CO (µg/m³)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Set the x-axis limit
plt.xlim(0, 600)

# Add a grid for the y-axis
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set global font sizes
plt.rcParams.update({
    'axes.titlesize': 42,  # Title size
    'axes.labelsize': 36,  # Label size
    'xtick.labelsize': 30,  # X-tick label size
    'ytick.labelsize': 30   # Y-tick label size
})

# Create the first figure with AQI and CO
fig1, axes1 = plt.subplots(1, 2, figsize=(20, 8))  # 1x2 grid for the first figure

# AQI bar graph
aqi_counts = merged_data['AQI'].value_counts().sort_index()
axes1[0].bar(aqi_counts.index, aqi_counts.values, color='skyblue', edgecolor='black')
axes1[0].set_title("Frequency of AQI Values")
axes1[0].set_xlabel("AQI Value")
axes1[0].set_ylabel("Frequency")
axes1[0].grid(axis='y', linestyle='--', alpha=0.5)

# CO histogram with KDE
sns.histplot(merged_data['CO'], bins=30, color='lightgreen', edgecolor='black', kde=True, ax=axes1[1])
axes1[1].set_title("Distribution of CO Levels")
axes1[1].set_xlabel("CO (µg/m³)")
axes1[1].set_ylabel("Frequency")
axes1[1].grid(axis='y', linestyle='--', alpha=0.5)
axes1[1].set_xlim(0, 500)

# Adjust layout for the first figure
plt.tight_layout()
plt.show()

# Create the second figure with NO2 and PM2.5
fig2, axes2 = plt.subplots(1, 2, figsize=(20, 8))  # 1x2 grid for the second figure

# NO2 histogram with KDE
sns.histplot(merged_data['NO2'], bins=30, color='gold', edgecolor='black', kde=True, ax=axes2[0])
axes2[0].set_title("Distribution of NO2 Levels")
axes2[0].set_xlabel("NO2 (µg/m³)")
axes2[0].set_ylabel("Frequency")
axes2[0].grid(axis='y', linestyle='--', alpha=0.5)
axes2[0].set_xlim(0, 30)

# PM2.5 histogram with KDE
sns.histplot(merged_data['PM2.5'], bins=30, color='red', edgecolor='black', kde=True, ax=axes2[1])
axes2[1].set_title("Distribution of PM2.5 Levels")
axes2[1].set_xlabel("PM2.5 (µg/m³)")
axes2[1].set_ylabel("Frequency")
axes2[1].grid(axis='y', linestyle='--', alpha=0.5)
axes2[1].set_xlim(0, 30)

# Adjust layout for the second figure
plt.tight_layout()
plt.show()
```
